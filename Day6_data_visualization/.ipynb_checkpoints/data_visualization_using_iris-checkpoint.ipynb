```python
# Data visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv("iris.csv")
```

```python
iris.info()
```

```python
iris.head()
```

```python
# To see how many no. of species are there
# and what kind of species are ther to deal with
iris['species'].value_counts()
```

```python
# Plotting data
# Here we are plotting sepal_length vs sepal_width 
# setosa - 'red'
# versicolor - 'blue'
# virginica - 'green'
for n in range(0,150):
    if iris['species'][n] == 'setosa':
        plt.scatter(iris['sepal_length'][n], iris['sepal_width'][n], color = 'red')
        plt.xlabel('sepal_length')
        plt.ylabel('sepal_width')
    elif iris['species'][n] == 'versicolor':
        plt.scatter(iris['sepal_length'][n], iris['sepal_width'][n], color = 'blue')
        plt.xlabel('sepal_length')
        plt.ylabel('sepal_width')
    elif iris['species'][n] == 'virginica':
        plt.scatter(iris['sepal_length'][n], iris['sepal_width'][n], color = 'green')
        plt.xlabel('sepal_length')
        plt.ylabel('sepal_width')
```

```python
# Here use the seaborn library
# Joint plots shows bivariate scatterplots
# And univariate histograms
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris)
```

```python
# Differntiating between species using facegrid
# Using Facetgrid function of seaborn
i = sns.FacetGrid(iris, hue = 'species', size = 5)
i = (i.map(plt.scatter, 'sepal_length', 'sepal_width').add_legend())
```

```python
# Here we plot the data with the regression/ boundary line
# Species are differentiated using hue
sns.lmplot(x = 'sepal_length', y = 'sepal_width', data = iris)
sns.lmplot(x = 'sepal_length', y = 'sepal_width', data = iris, hue = 'species')
# sns.lmplot(x = 'sepal_length', y = 'sepal_width', data = iris)
```

```python
# This graph is same as above but plotting the species separately
sns.lmplot(x = 'sepal_length', y = 'sepal_width', data = iris, hue = 'species', col = 'species')
```

```python
# Plottign data in strip
plt.subplot(2,2,1)
sns.stripplot(x = 'species', y = 'sepal_length', data = iris, jitter = True)
plt.subplot(2,2,2)
sns.stripplot(x = 'species', y = 'sepal_width', data = iris, jitter = True)
plt.subplot(2,2,3)
sns.stripplot(x = 'species', y = 'petal_length', data = iris, jitter = True)
plt.subplot(2,2,4)
sns.stripplot(x = 'species', y = 'petal_width', data = iris, jitter = True)
```

```python
# Plotting the features using boxes
plt.style.use('ggplot')
plt.subplot(2,2,1)
sns.boxplot(x = 'species', y = 'sepal_length', data = iris)
plt.subplot(2,2,2)
sns.boxplot(x = 'species', y = 'sepal_width', data = iris)
plt.subplot(2,2,3)
sns.boxplot(x = 'species', y = 'petal_length', data = iris)
plt.subplot(2,2,4)
sns.boxplot(x = 'species', y = 'petal_width', data = iris)
```

```python
# Representing data using violin form
plt.style.use('ggplot')
plt.subplot(2,2,1)
sns.violinplot(x = 'species', y = 'sepal_length', data = iris)
plt.subplot(2,2,2)
sns.violinplot(x = 'species', y = 'sepal_width', data = iris)
plt.subplot(2,2,3)
sns.violinplot(x = 'species', y = 'petal_length', data = iris)
plt.subplot(2,2,4)
sns.violinplot(x = 'species', y = 'petal_width', data = iris)
```

```python
# Plotting pairplots 
# Using pairplot we can relate each  pair of feature with other features
# Here we can use KDE form insteaad of histogram by changing 'kind' to 'kde'
sns.pairplot(iris, hue = 'species')
```
